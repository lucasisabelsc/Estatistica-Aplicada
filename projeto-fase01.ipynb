# Pergunta 4: 
Ao menos dois gráficos criativos para ajudar na visualização de resultados, conforme orientações dadas em aula pelo professor Hélio;

## Altura de Goleiros vs Não Goleiros

```python
sns.boxplot(data=df, x='Goleiro', y='height_cm')
plt.title('Distribuição da Altura por Posição de Goleiro')
plt.xlabel('É Goleiro (1 = Sim, 0 = Não)')
plt.ylabel('Altura (cm)')
plt.show()
```

Observa-se que a variabilidade da altura para não goleiros é muito maior do que para os goleiros. Além disso, a mediana dos goleiros está a acima do percentil 75% dos não goleiros.

## Gráfico de Overall Rating por Nacionalidade

```python
top_nations = df['nationality'].value_counts()
top_nations = top_nations[top_nations > 10].index

media_rating = (
    df[df['nationality']
    .isin(top_nations)]
    .groupby('nationality')['overall_rating']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

media_rating
```

```python
sns.barplot(
    data=media_rating, 
    x='overall_rating', 
    y='nationality', 
    palette='viridis',
    hue='nationality',
    legend=False,
    )
plt.title('Top 10 Nacionalidades com Maior Média de Overall Rating')
plt.xlabel('Média do Overall Rating')
plt.ylabel('Nacionalidade')
plt.tight_layout()
plt.show()
```

# Pergunta 5:
Um histograma ou gráfico de barra para verificação de adequabilidade do modelo probabilístico para a variável escolhida

## Histograma Altura dos Jogadores

```python
sns.histplot(
    data=df, 
    x='height_cm', 
    kde=True, 
    color='skyblue', 
    edgecolor='black')
plt.title('Distribuição da Altura dos Jogadores')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Observa-se que a curva de densidade apresentada se difere muito da curva normal. Temos alguns "buracos" entre 170 e 180cm e uma frequência muito alta próximo a 155cm - gerando dois picos de densidade.
